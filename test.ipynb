```python
import os
import pandas as pd
from sqlalchemy import create_engine
from dotenv import load_dotenv, find_dotenv

# 1. FORZAR la b√∫squeda del .env
# find_dotenv() busca el archivo en carpetas superiores si no est√° en la actual
archivo_env = find_dotenv()

if archivo_env == "":
    print("‚ùå ERROR CR√çTICO: No se encontr√≥ el archivo .env")
    print("Aseg√∫rate de haber creado un archivo llamado exactamente '.env' (sin .txt al final)")
else:
    load_dotenv(archivo_env)
    print(f"‚úÖ Archivo .env encontrado en: {archivo_env}")

# 2. Verificar que las variables se cargaron (Depuraci√≥n)
db_port = os.getenv("DB_PORT")
print(f"Puerto detectado: {db_port}") # Si esto imprime 'None', el archivo est√° mal escrito

# 3. Conectar (Solo si tenemos puerto)
if db_port:
    user = os.getenv("DB_USER")
    password = os.getenv("DB_PASSWORD")
    host = os.getenv("DB_HOST")
    dbname = os.getenv("DB_NAME")
    
    # Aseguramos que el puerto sea un string para la concatenaci√≥n
    connection_str = f"postgresql://{user}:{password}@{host}:{db_port}/{dbname}"
    
    try:
        engine = create_engine(connection_str)
        print("‚úÖ Conexi√≥n configurada. Probando query...")
        # pd.read_sql("SELECT 1", engine) # Descomentar para probar
    except Exception as e:
        print(f"‚ùå Error de conexi√≥n: {e}")
else:
    print("‚ùå Detenido: Revisa tu archivo .env, falta la variable DB_PORT")
```

```python
from sqlalchemy import inspect

# Crear el inspector
inspector = inspect(engine)

# Obtener los nombres de las tablas
mis_tablas = inspector.get_table_names()

print("üìã Tablas disponibles en tu base de datos:")
print(mis_tablas)
```

```python
nombre_tabla = 'productos'  # <--- Pon aqu√≠ el nombre que te sali√≥ en el paso 1

# Leemos la tabla completa
df = pd.read_sql(f"SELECT * FROM {nombre_tabla}", engine)

print(df.head())
```
